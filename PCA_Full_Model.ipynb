```python
#next we need to scale the lab values 
lab_cols = ['PRSODM', 'PRBUN', 'PRCREAT', 'PRWBC', 'PRHCT', 'PRPLATE']

def scale_labs(df, lab_cols):
    # create a DataFrame with only the lab columns
    lab_df = df[lab_cols]

    # replace missing values (-99) with NaN
    lab_df = lab_df.replace(-99, np.nan)

    # impute missing values with the median
    imputer = SimpleImputer(strategy='median')
    
    #transform lab values 
    lab_df = imputer.fit_transform(lab_df)

    # create a StandardScaler object
    scaler = StandardScaler()

    # fit the scaler to the lab data
    scaler.fit(lab_df)

    # transform the lab data
    scaled_data = scaler.transform(lab_df)

    # create a new DataFrame with the scaled lab data
    scaled_lab_df = pd.DataFrame(scaled_data, columns=lab_cols)

    # create a new DataFrame that contains both the scaled lab columns and the other columns from the input DataFrame
    new_df = pd.concat([df.drop(lab_cols, axis=1), scaled_lab_df], axis=1)

    return new_df


scaled_df  = scale_labs(full_df, lab_cols)
```

```python

#convert these two to binary 
display(scaled_df['EMERGNCY'].value_counts())
display(scaled_df['INOUT'].value_counts())
```

```python
#convert EMERGENCY and INOUT
def convert_binary(df): 
    encoding_EMERGENCY = {'No': 0, 'Yes': 1}
    encoding_INOUT = {'Inpatient': 1, 'Outpatient': 0}

    df['EMERGNCY'] = df['EMERGNCY'].map(encoding_EMERGENCY)
    df['INOUT'] = df['INOUT'].map(encoding_INOUT)
    df_encoded = df 
    return df_encoded

df_encoded = convert_binary(scaled_df)
```

```python
#convert these two to binary 
display(df_encoded['EMERGNCY'].value_counts())
display(df_encoded['INOUT'].value_counts())
```

```python
def split_data(df): 
    X = df.drop(labels = ['READMISSION30D'], axis = 1) 
    y = pd.DataFrame(df['READMISSION30D']) 
    X_train,X_test,y_train,y_test = train_test_split(X, y, stratify = y, test_size= .2, random_state= 0 )
    return X_train,X_test,y_train,y_test

X_train,X_test,y_train,y_test = split_data(df_encoded)
display(X_train.info()) 
display(X_test.info())
display(y_train.info())
display(y_test.info())
```

```python
# Train a PCA model with all features
pca_all = PCA()
pca_all.fit(X_train)
```

```python
# Get the top 20 principal components
pcs_top20 = pca_all.transform(X_train)[:, :20]

# Train a PCA model using the top 20 principal components
pca_top20 = PCA()
pca_top20.fit(pcs_top20)
```

```python
# Transform the testing set using the PCA model with all features
pcs_test_all = pca_all.transform(X_test)

# Transform the testing set using the PCA model with top 20 principal components
pcs_test_top20 = pcs_test_all[:, :20]

# Reconstruct the testing set using the PCA model with top 20 principal components
X_test_top20_reconstructed = pca_top20.inverse_transform(pcs_test_top20)
```

```python
# Get the predicted probabilities for the reconstructed testing set
probs_top20 = np.exp(X_test_top20_reconstructed[:, 0]) / (1 + np.exp(X_test_top20_reconstructed[:, 0]))

# Unexpectedly an NaN created by the above operation
imputer1 = SimpleImputer(strategy='median')
probs_top20 = probs_top20.reshape(-1, 1)
probs_top20 = imputer1.fit_transform(probs_top20)

```

```python
# Calculate the false positive rate and true positive rate for all thresholds
fpr_top20, tpr_top20, _ = roc_curve(y_test, probs_top20)

# Calculate the AUC score for the PCA model with top 20 principal components
auc_top20 = roc_auc_score(y_test, probs_top20)

# Plot the ROC curve for the PCA model with top 20 principal components
roc_display_top20 = RocCurveDisplay(fpr=fpr_top20, tpr=tpr_top20, roc_auc=auc_top20, estimator_name='PCA (Top 20 principal components)')
roc_display_top20.plot()
plt.text(0.5, 0.1, f"AUC = {auc_top20:.3f}", ha='center', fontsize=12)
plt.plot([0,1],[0,1],linestyle='--',label='Random guessing')
plt.legend()
plt.show()

# Reconstruct the testing set using the PCA model with all principal components
X_test_all_reconstructed = pca_all.inverse_transform(pcs_test_all)

# Get the predicted probabilities for the reconstructed testing set using the PCA model with all principal components
probs_all = np.exp(X_test_all_reconstructed[:, 0]) / (1 + np.exp(X_test_all_reconstructed[:, 0]))

# Calculate the false positive rate and true positive rate for all thresholds
fpr_all, tpr_all, _ = roc_curve(y_test, probs_all)

# Calculate the AUC score for the PCA model with all principal components
auc_all = roc_auc_score(y_test, probs_all)

# Plot the ROC curve for the PCA model with all principal components
roc_display_all = RocCurveDisplay(fpr=fpr_all, tpr=tpr_all, roc_auc=auc_all, estimator_name='PCA (All principal components)')
roc_display_all.plot()
plt.text(0.5, 0.1, f"AUC = {auc_all:.3f}", ha='center', fontsize=12)
plt.plot([0,1],[0,1],linestyle='--',label='Random guessing')
plt.legend()
plt.show()
```
